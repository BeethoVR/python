```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage import color, io
```

```python
img = io.imread('/kaggle/input/cnn-data-sources/yo.png')
```

```python
plt.imshow(img)
```

```python
img_gray = color.rgb2gray(img)
```

```python
plt.imshow(img_gray)
```

```python
#kernel = np.array([[-1,0,1], 
#                   [-1,0,1], 
#                   [-1,0,1]]) # Filtro para bordes Verticales

kernel = np.array([[0,0,0], 
                   [0,1,0], 
                   [0,0,0]]) # Filtro para Identidad
```

```python
img_new = nd.convolve(img_gray, kernel)
```

```python
plt.imshow(img_new)
```

```python
fig, axes = plt.subplots(1,2, figsize=(15,10))
axes[0].imshow(img_gray, cmap=plt.cm.gray)
axes[0].set_title('Imagen Original')
axes[0].axis('off')

axes[1].imshow(img_new, cmap=plt.cm.gray)
axes[1].set_title('Convolución')
axes[1].axis('off')
```
